```python
import pandas as pd

df = pd.read_csv('bbc_news.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 7'], axis=1)
```

```python
df
```

```python
df = df[~df['guid'].str.contains("sport", na=False, case=False)]
df = df[~df['guid'].str.contains("health", na=False, case=False)]
df = df[~df['guid'].str.contains("london", na=False, case=False)]
df = df[~df['guid'].str.contains("england", na=False, case=False)]
df = df[~df['guid'].str.contains("scotland", na=False, case=False)]
df = df[~df['guid'].str.contains("wales", na=False, case=False)]
df = df[~df['guid'].str.contains("Australia", na=False, case=False)]
df = df[~df['guid'].str.contains("entertainment", na=False, case=False)]
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'], format='%a, %d %b %Y %H:%M:%S %Z', errors='raise')
```

```python
df['date'] = df['date'].dt.strftime('%Y/%m/%d')
```

```python
df.to_csv('test.csv',index=False)
```
