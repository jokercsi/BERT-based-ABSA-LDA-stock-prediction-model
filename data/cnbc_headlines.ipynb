```python
import pandas as pd

df = pd.read_csv('cnbc_headlines.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna(axis=0)
```

```python
df['Time']
```

```python
df['Time'] = df['Time'].str.lstrip()
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%I:%M  %p %Z %a, %d %b %Y', errors='raise')
```

```python
df['Time']
```

```python
df
```

```python
df['Time'] = df['Time'].dt.strftime('%Y-%m-%d')
```

```python

df = df.sort_values(by='Time')
```
