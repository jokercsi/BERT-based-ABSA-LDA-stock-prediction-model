```python
# 필요한 라이브러리 가져오기
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
from datetime import datetime
```

```python
# 대상 사이트 : investing.com
url = "https://www.investing.com/news/stock-market-news"

# 프로토콜 status 확인
resp = requests.get(url)
print(resp)
```

```python
# 가져온 사이트 코드 
soup = BeautifulSoup(resp.content, 'lxml')
print(soup)
```

```python
url_list =[]
# url_list = soup.find_all('article', "js-article-item")

# Article 태그의 js-article-item클래스만 가져오기
for i in soup.find_all('article', "js-article-item"):

	# 광고태그는 제거하기
	# i['class'] => i의 클래스 이름 가져오기 
  if "sponsoredArticle" in i['class']:
    continue
     
  else:
    # 여러개 a태그 중에 첫번째 태그만 가져오기
    url_list.append(i.find('a'))


# a 태그 가져오기
# url_list = soup.find('div', 'largeTitle').find_all('a')
print(url_list)
```

```python
news_url_list = []

for url in url_list:
  news_url_list.append('https://www.investing.com' + url['href'])
  #print(url['href'])
```

```python
# 링크가 주어지면, 기사 제목과 본문을 크롤링하는 코드
def get_news_by_url(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'} # 각자 본인의 User-Agent 정보 입력 
    res = requests.get(url, headers = headers)
    bs = BeautifulSoup(res.content, 'html.parser')

    #print(bs.find('h1','articleHeader'))

    #제목
    title = bs.find('h1','articleHeader').string
    title = re.sub('\t|\r|\n| ', '', title)
    #본문
    content = []
    for i in bs.find('div','articlePage').find_all('p') :
      content.append(i.text)

    content = "".join(content)
    
    return  title + content
```

```python
#url을 통해 각 뉴스기사 페이지의 본문을 가져온다!
news_content = []
for url in news_url_list :
    news_content.append(get_news_by_url(url))
```

```python
for i in news_content:
  print(i)
```
